```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../1-Data/heart.csv")
print('Shape',df.shape)
```

1. Remove the RestingBP outliers
2. Split train/test
3. Change the cholesterol = 0
4. Encode categorical

```python
df_model = df.copy()
# 1. Remove RestingBP outliers
df_model = df_model[df_model["RestingBP"] > 0]

print("Shape after removing RestingBP outliers:")
print(df_model.shape)
```

```python
# 2. Encode categorical
df_model["Sex"] = (df_model["Sex"] == "M").astype(int)

df_model["ExerciseAngina"] = (df_model["ExerciseAngina"] == "Y").astype(int)

# Ordinal encoding
df_model["ST_Slope"] = df_model["ST_Slope"].map({
    "Down": 0,
    "Flat": 1,
    "Up": 2
})

# One-hot encoding
df_model = pd.get_dummies(
    df_model,
    columns=["ChestPainType", "RestingECG"],
    drop_first=True,
    dtype=int 
)

print("\nShape after encoding:")
print(df_model.shape)
```

```python
# 3. Split the dataset
df_train, df_test = train_test_split(
    df_model,
    test_size=0.2,
    random_state=42,
    stratify=df_model['HeartDisease']
)

df_train.to_csv("../1-Data/train_data.csv", index = False)
df_test.to_csv("../1-Data/test_data.csv", index=False)
```

```python
# 4. Change the cholesterol = 0 
train_df = pd.read_csv("../1-Data/train_data.csv")

def add_cholesterol_missing_flag(df):
    df = df.copy()
    df["Cholesterol_missing"] = (df["Cholesterol"] == 0).astype(int)
    return df


def impute_cholesterol(df, median_value=None):
    df = df.copy()
    df["Cholesterol"] = df["Cholesterol"].replace(0, np.nan)
    if median_value is None:
        median_value = df["Cholesterol"].median()
    df["Cholesterol"] = df["Cholesterol"].fillna(median_value)
    return df, median_value

train_df = add_cholesterol_missing_flag(train_df)
train_df_processed, median_chol = impute_cholesterol(train_df)
print("\nTrain missing values after imputation:")
print(train_df_processed["Cholesterol"].isnull().sum())
train_df_processed.to_csv("../1-Data/train_processed.csv", index=False)

# Apply the same train median to the test set so both datasets have the same feature shape

test_df = pd.read_csv("../1-Data/test_data.csv")
test_df = add_cholesterol_missing_flag(test_df)
test_df_processed, _ = impute_cholesterol(test_df, median_value=median_chol)
print("\nTest missing values after imputation:")
print(test_df_processed["Cholesterol"].isnull().sum())
test_df_processed.to_csv("../1-Data/test_processed.csv", index=False)
```
