## 0. Import Thu Vien

```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, classification_report,
                             confusion_matrix, roc_auc_score, roc_curve)

sns.set_theme(style='whitegrid')
COLORS = ['#3DC5A8', '#E8593C']   # Normal vs Disease
SEED   = 42
print('Libraries loaded.')
```

---
## Buoc 1 — Tong Quan Du Lieu
Kiem tra kich thuoc, kieu du lieu, thong ke mo ta.


```python
import os
_paths = ['../input/heart-failure-prediction/heart.csv',
          '../1-Data/train_data.csv', 'heart.csv']
path = next((p for p in _paths if os.path.exists(p)), None)
if path is None: raise FileNotFoundError('Khong tim thay file. Sua path.')

df = pd.read_csv(path)
print(f'Shape  : {df.shape}')
print(f'Columns: {list(df.columns)}')
df.head()
```

## Buoc 2 — Kiem Tra Chat Luong Du Lieu
Kiem tra: null values, duplicate rows, gia tri khong hop le.


```python
print('=== Null Values ===')
print(df.isnull().sum())

print(f'\n=== Duplicate Rows === {df.duplicated().sum()}')

print('\n=== Gia tri 0 khong hop le ===')
print(f'Cholesterol == 0: {(df["Cholesterol"]==0).sum()} records')
print(f'RestingBP   == 0: {(df["RestingBP"]==0).sum()} records')
print('=> Cholesterol=0 khong the xay ra o nguoi song -> missing data')
print('=> Can xu ly truoc khi train model (Buoc 6)')
```

```python
# Kiem tra phan bo gia tri 0 theo class
n0 = df[df['Cholesterol']==0].groupby('HeartDisease').size()
pct = (n0 / df.groupby('HeartDisease').size() * 100).round(1)
print('Cholesterol=0 phan bo theo HeartDisease:')
print(pd.DataFrame({'so_luong': n0, 'phan_tram_%': pct}))
print('\n=> Nhieu hon o nhom Disease -> MNAR (Missing Not At Random)')
print('=> Dung median theo tung nhom de impute, khong dung global median')
```

## Buoc 3 — Phan Tich Phan Phoi Tung Bien (Univariate)


```python
# --- Bien phan loai (Categorical) ---
cat_cols = ['Sex','ChestPainType','FastingBS','RestingECG','ExerciseAngina','ST_Slope']

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
fig.suptitle('Phan phoi cac bien phan loai', fontsize=14, fontweight='bold')

for ax, col in zip(axes.flatten(), cat_cols):
    order = sorted(df[col].astype(str).unique())
    sns.countplot(x=col, data=df, order=order, color='#5B9BD5',
                  edgecolor='black', ax=ax)
    for p in ax.patches:
        pct = p.get_height() / len(df) * 100
        if p.get_height() > 0:
            ax.annotate(f'{p.get_height()}\n({pct:.0f}%)',
                        (p.get_x()+p.get_width()/2, p.get_height()),
                        ha='center', va='bottom', fontsize=8)
    ax.set_title(col); ax.set_xlabel('')

plt.tight_layout()
plt.show()
```

```python
# --- Bien so (Numerical) ---
num_cols = ['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
fig.suptitle('Phan phoi cac bien so', fontsize=14, fontweight='bold')

for ax, col in zip(axes.flatten(), num_cols):
    sns.histplot(df[col], kde=True, bins=30, color='#E8593C',
                 edgecolor='white', ax=ax)
    ax.axvline(df[col].mean(),   color='blue', ls='--', lw=1.5, label='Mean')
    ax.axvline(df[col].median(), color='green',ls=':',  lw=1.5, label='Median')
    ax.set_title(f'{col}  (skew={df[col].skew():.2f})')
    ax.legend(fontsize=7)

fig.delaxes(axes.flatten()[-1])
plt.tight_layout()
plt.show()

print('Nhan xet skewness:')
for col in num_cols:
    s = df[col].skew()
    note = 'Right-skewed (lech phai)' if s>0.5 else 'Left-skewed' if s<-0.5 else 'Gan chuan'
    print(f'  {col:<14}: {s:+.2f}  -> {note}')
```

## Buoc 4 — Phan Tich Bien vs Target (Bivariate)
Xem moi bien anh huong the nao den HeartDisease.


```python
# --- Target distribution ---
counts = df['HeartDisease'].value_counts()
ratios = df['HeartDisease'].value_counts(normalize=True)
print('Phan bo Target (HeartDisease):')
print(pd.DataFrame({'so_luong': counts, 'ti_le': ratios.map('{:.1%}'.format)}))

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Phan bo HeartDisease', fontweight='bold')

axes[0].pie([counts[1]/len(df)*100, counts[0]/len(df)*100],
            labels=['Disease','Normal'], autopct='%1.1f%%',
            colors=COLORS[::-1], startangle=90,
            wedgeprops={'edgecolor':'black'})
axes[0].set_title('Ti le')

bars = axes[1].bar(['Normal (0)','Disease (1)'], counts.values,
                   color=COLORS, edgecolor='black', width=0.5)
for bar, n in zip(bars, counts.values):
    axes[1].text(bar.get_x()+bar.get_width()/2, bar.get_height()+5,
                 str(n), ha='center', fontweight='bold')
axes[1].set_title('So luong')
plt.tight_layout(); plt.show()
```

```python
# --- Categorical vs Target ---
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Categorical Features vs HeartDisease', fontsize=13, fontweight='bold')

for ax, col in zip(axes.flatten(), cat_cols):
    order = sorted(df[col].astype(str).unique())
    sns.countplot(x=col, hue='HeartDisease', data=df, order=order,
                  palette={0:COLORS[0],1:COLORS[1]}, edgecolor='black', ax=ax)

    # Disease rate % tren moi category
    ax2 = ax.twinx()
    rate = df.groupby(col)['HeartDisease'].mean().reindex(order).values
    ax2.plot(range(len(order)), rate, color='navy',
             marker='o', lw=2, markersize=7, zorder=5)
    for xi, val in enumerate(rate):
        if not np.isnan(val):
            ax2.annotate(f'{val:.0%}', (xi, val+0.05),
                         ha='center', fontsize=9, color='navy', fontweight='bold')
    ax2.set_ylim(0, 1.15); ax2.set_ylabel('Disease rate %', fontsize=8)
    ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x,_: f'{x:.0%}'))
    ax.set_title(col); ax.set_xlabel('')
    ax.legend(['Normal','Disease'], fontsize=8)

plt.tight_layout(); plt.show()
```

```python
# --- Numerical vs Target ---
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Numerical Features vs HeartDisease', fontsize=13, fontweight='bold')

for ax, col in zip(axes.flatten(), num_cols):
    sns.boxplot(data=df, x='HeartDisease', y=col,
                hue='HeartDisease',
                palette={0:COLORS[0],1:COLORS[1]},
                legend=False, ax=ax)
    for cv, xp in [(0,0),(1,1)]:
        med = df[df['HeartDisease']==cv][col].median()
        ax.text(xp+0.12, med, f'  {med:.0f}',
                va='center', fontsize=9, fontweight='bold')
    ax.set_title(col)
    ax.set_xticks([0,1]); ax.set_xticklabels(['Normal','Disease'])
    ax.set_xlabel('')

fig.delaxes(axes.flatten()[-1])
plt.tight_layout(); plt.show()
```

---
## Buoc 5 — Ma Tran Tuong Quan (Correlation)
Encode categorical dung cach, sau do tinh tuong quan Pearson.

> **Luu y:** Chi encode binary (2 gia tri) va ordinal (co thu tu) bang so.
> Nominal nhieu gia tri (ChestPainType, RestingECG) -> dung One-Hot hoac Cramer's V rieng.


```python
df_corr = df.copy()

# Encode dung: binary + ordinal
df_corr['Sex_enc']            = (df_corr['Sex'] == 'M').astype(int)
df_corr['ExerciseAngina_enc'] = (df_corr['ExerciseAngina'] == 'Y').astype(int)
df_corr['ST_Slope_enc']       = df_corr['ST_Slope'].map({'Down':0,'Flat':1,'Up':2})

corr_cols = ['Age','RestingBP','Cholesterol','MaxHR','Oldpeak',
             'FastingBS','Sex_enc','ExerciseAngina_enc','ST_Slope_enc','HeartDisease']

corr = df_corr[corr_cols].corr()
rename = {'Sex_enc':'Sex','ExerciseAngina_enc':'ExerciseAngina','ST_Slope_enc':'ST_Slope'}
corr = corr.rename(columns=rename, index=rename)

plt.figure(figsize=(10, 7))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, square=True, vmin=-1, vmax=1,
            linewidths=0.4, annot_kws={'size':9})
plt.title('Ma tran tuong quan\n'
          '(Chi bao gom bien binary/ordinal da encode dung cach)',
          fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout(); plt.show()

print('Tuong quan voi HeartDisease (sap xep theo |r|):')
corr_target = corr['HeartDisease'].drop('HeartDisease').abs().sort_values(ascending=False)
print(corr_target.to_string())
```

---
## Buoc 6 — Lam Sach & Chuan Bi Du Lieu
3 buoc can thiet truoc khi train Logistic Regression:
1. Xu ly Cholesterol=0 (missing data)
2. Encode categorical
3. Scale numerical (bat buoc voi Logistic Regression)


```python
df_model = df.copy()

# --- 1. Xu ly Cholesterol=0 ---
# Tao flag de giu thong tin 'ban dau bi missing'
df_model['Cholesterol_missing'] = (df_model['Cholesterol']==0).astype(int)
# Replace 0 -> NaN -> impute bang median theo nhom
df_model['Cholesterol'] = df_model['Cholesterol'].replace(0, np.nan)
df_model['Cholesterol'] = (df_model.groupby('HeartDisease')['Cholesterol']
                           .transform(lambda x: x.fillna(x.median())))
print(f'Cholesterol null sau impute: {df_model["Cholesterol"].isnull().sum()}')

# --- 2. Encode categorical ---
# Binary
df_model['Sex']            = (df_model['Sex']=='M').astype(int)
df_model['ExerciseAngina'] = (df_model['ExerciseAngina']=='Y').astype(int)
df_model['ST_Slope']       = df_model['ST_Slope'].map({'Down':0,'Flat':1,'Up':2})
# One-Hot cho nominal
df_model = pd.get_dummies(df_model,
                           columns=['ChestPainType','RestingECG'],
                           drop_first=True)

print(f'Shape sau encode: {df_model.shape}')
print(f'Columns: {list(df_model.columns)}')
```

```python
# --- 3. Tach X, y va Train/Test split ---
X = df_model.drop('HeartDisease', axis=1)
y = df_model['HeartDisease']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=SEED)

print(f'X_train: {X_train.shape} | X_test: {X_test.shape}')
print(f'Train Disease rate: {y_train.mean()*100:.1f}%')
print(f'Test  Disease rate: {y_test.mean()*100:.1f}%')

# --- 4. Scale (BAT BUOC voi Logistic Regression) ---
scaler  = StandardScaler()
X_train = scaler.fit_transform(X_train)   # fit chi tren train
X_test  = scaler.transform(X_test)        # chi transform test (khong fit lai)
print('\nStandardScaler applied.')
print('=> Fit chi tren X_train de tranh data leakage')
```

---
## Buoc 7 — Xay Dung Mo Hinh Logistic Regression


```python
# --- Train model ---
model = LogisticRegression(
    C=1.0,              # Regularization strength (1/lambda)
    penalty='l2',       # L2 Ridge Regularization
    max_iter=1000,
    random_state=SEED
)
model.fit(X_train, y_train)
print('Model trained.')

# --- Cross Validation (5-fold) ---
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
cv_scores = cross_val_score(model, X_train, y_train,
                             cv=cv, scoring='roc_auc')
print(f'\nCV ROC-AUC (5-fold): {cv_scores.mean():.4f} +/- {cv_scores.std():.4f}')
print(f'Scores: {[round(s,4) for s in cv_scores]}')
```

```python
# --- Coefficients (feature importance cua Logistic Regression) ---
feat_names = X.columns.tolist()
coef_df    = pd.DataFrame({'Feature': feat_names,
                            'Coefficient': model.coef_[0]})
coef_df    = coef_df.reindex(
    coef_df['Coefficient'].abs().sort_values(ascending=False).index
)

plt.figure(figsize=(10, 7))
colors_bar = [COLORS[1] if c > 0 else COLORS[0] for c in coef_df['Coefficient']]
bars = plt.barh(coef_df['Feature'], coef_df['Coefficient'],
                color=colors_bar, edgecolor='black', height=0.6)
plt.axvline(0, color='black', lw=0.8)
plt.title('Logistic Regression Coefficients (L2 Regularization)\n'
          'Do > 0 = tang xac suat Disease | Xanh < 0 = giam xac suat',
          fontweight='bold')
plt.xlabel('Coefficient value')
plt.tight_layout()
plt.show()
print('Top 5 features anh huong TANG disease risk:')
print(coef_df[coef_df['Coefficient']>0].head(5)[['Feature','Coefficient']].to_string(index=False))
print('\nTop 5 features anh huong GIAM disease risk:')
print(coef_df[coef_df['Coefficient']<0].head(5)[['Feature','Coefficient']].to_string(index=False))
```

---
## Buoc 8 — Danh Gia Mo Hinh


```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print('='*50)
print('KET QUA TREN TAP TEST')
print('='*50)
print(f'Accuracy : {accuracy_score(y_test, y_pred):.4f}')
print(f'ROC-AUC  : {roc_auc_score(y_test, y_prob):.4f}')
print(f'\nClassification Report:')
print(classification_report(y_test, y_pred,
      target_names=['Normal (0)','Disease (1)']))
```

```python
# Confusion Matrix + ROC Curve
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('Danh Gia Mo Hinh Logistic Regression', fontweight='bold')

# --- Confusion Matrix ---
cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = cm.ravel()
lbs = np.array([[f'TN={tn}\n(Du doan dung Normal)',
                  f'FP={fp}\n(Bao nham Disease)'],
                 [f'FN={fn}\n(Bo sot benh nhan!)',
                  f'TP={tp}\n(Du doan dung Disease)']])
sns.heatmap(cm, annot=lbs, fmt='', cmap=['#3DC5A8','#E8593C'],
            linewidths=0.5, ax=axes[0], cbar=False)
axes[0].set_title('Confusion Matrix')
axes[0].set_xlabel('Predicted'); axes[0].set_ylabel('Actual')
axes[0].set_xticklabels(['Normal','Disease'])
axes[0].set_yticklabels(['Normal','Disease'])

# --- ROC Curve ---
fpr, tpr, _ = roc_curve(y_test, y_prob)
auc_score   = roc_auc_score(y_test, y_prob)
axes[1].plot(fpr, tpr, color=COLORS[1], lw=2,
             label=f'Logistic Regression (AUC = {auc_score:.3f})')
axes[1].plot([0,1],[0,1],'k--', lw=1, label='Random (AUC=0.500)')
axes[1].fill_between(fpr, tpr, alpha=0.1, color=COLORS[1])
axes[1].set(xlabel='False Positive Rate (1 - Specificity)',
            ylabel='True Positive Rate (Recall/Sensitivity)',
            title='ROC Curve', xlim=(0,1), ylim=(0,1.02))
axes[1].legend(fontsize=10)

plt.tight_layout(); plt.show()

print(f'FN = {fn}: so benh nhan bi benh nhung model bao la Normal (nguy hiem nhat)')
print(f'FP = {fp}: so nguoi khoe manh bi bao la Disease (kham them, chap nhan duoc)')
```

```python
# --- So sanh L1 vs L2 Regularization ---
print('So sanh L1 (Lasso) vs L2 (Ridge) Regularization:')
print(f'  {"C":<8} {"L1 Acc":>10} {"L1 AUC":>10} {"L2 Acc":>10} {"L2 AUC":>10}')
print('-'*55)
for C in [0.01, 0.1, 1.0, 10.0, 100.0]:
    m_l1 = LogisticRegression(C=C, penalty='l1', solver='liblinear',
                               max_iter=1000).fit(X_train, y_train)
    m_l2 = LogisticRegression(C=C, penalty='l2', solver='lbfgs',
                               max_iter=1000).fit(X_train, y_train)
    acc1 = accuracy_score(y_test, m_l1.predict(X_test))
    auc1 = roc_auc_score(y_test, m_l1.predict_proba(X_test)[:,1])
    acc2 = accuracy_score(y_test, m_l2.predict(X_test))
    auc2 = roc_auc_score(y_test, m_l2.predict_proba(X_test)[:,1])
    print(f'  {C:<8} {acc1:>10.4f} {auc1:>10.4f} {acc2:>10.4f} {auc2:>10.4f}')
print('\n=> C lon = Regularization yeu (mo hinh phuc tap hon)')
print('=> C nho = Regularization manh (mo hinh don gian hon, tranh overfit)')
```
